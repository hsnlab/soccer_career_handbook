# Normal test

```python

k2, p = stats.normaltest(tr[1]['lnvalue'].values)
print('Statistics=%.3f, p=%.3f' % (k2, p))
alpha = 1e-3

if p < alpha:  # null hypothesis: x comes from a normal distribution
    print("The null hypothesis can be rejected")
else:
    print("The null hypothesis cannot be rejected")
plt.hist(tr[1]['lnvalue'].values)
```

```python
for i in range(10):
    year=i+2012
    stat, p = normaltest(va[i]['lnvalue'].values)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print(str(year) + ' Fifa data looks normal (fail to reject H0)')
    else:
        print(str(year) + ' Fifa data does not look normal (reject H0)')
    print('----------')

for i in range(10):
    year=i+2012
    stat, p = normaltest(tr[i]['lnvalue'].values)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print(str(year) + ' Transfermarkt data looks normal (fail to reject H0)')
    else:
        print(str(year) + ' Transfermarkt data does not look normal (reject H0)')
    print('----------')
```

# MannWhitney

```python
for i in range(10):
    year=i+2012
    stat, p = mannwhitneyu(tr[i]['lnvalue'], va[i]['lnvalue'])
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.01
    if p > alpha:
        print(str(year) +' Same distribution (fail to reject H0)')
    else:
        print(str(year) +' Different distribution (reject H0)')
    print('----------')
```

# Kruskal

```python
for i in range(10):
    year=i+2012
    stat, p = kruskal(tr[i]['lnvalue'], va[i]['lnvalue'])
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.01
    if p > alpha:
        print(str(year) +' Same distribution (fail to reject H0)')
    else:
        print(str(year) +' Different distribution (reject H0)')
    print('----------')
```
